```python
albums_artist = pd.read_sql_query("SELECT AlbumId, Title, a.Name \
                           FROM albums \
                           LEFT JOIN artists as a \
                           ON a.ArtistId = albums.ArtistId", conn)
albums_artist
```

```python
albums_tracks = pd.read_sql_query("SELECT * \
                                 FROM albums \
                                 RIGHT JOIN tracks as t \
                                 ON t.AlbumId = albums.AlbumId", conn)
albums_tracks
```

```python
tracks_albums_genre = pd.read_sql_query("""
    SELECT tracks.*, a.Title, g.Name AS Genre_Name
    FROM tracks
    LEFT JOIN albums AS a ON a.AlbumId = tracks.AlbumId
    LEFT JOIN genres AS g ON g.GenreId = tracks.GenreId
""", conn)
```

```python
tracks_albums_genre.tail()
```

```python
genre_counts = tracks_albums_genre['Genre_Name'].value_counts()
genre_counts
```

```python
composer = tracks_albums_genre['Composer']
```

```python
price=tracks_albums_genre.groupby('Composer')['UnitPrice'].mean()
```

```python
price_99 = price.loc[price == 0.99]
print(price_99)
```

```python
top_cust = pd.read_sql_query("SELECT CustomerId, SUM(Total)  as TotalValue, \
                              COUNT(InvoiceId) as Purchases \
                              FROM INVOICES \
                              GROUP BY CustomerId \
                              ORDER BY TotalValue DESC \
                              ", con=conn, index_col='CustomerId')
top_cust
```

```python
top_genre = pd.read_sql_query("SELECT genres.GenreId, genres.Name, \
                               SUM(invoice_items.quantity * invoice_items.unitprice) AS Total, \
                               SUM(invoice_items.quantity) AS Jumlah_Pembelian \
                               FROM tracks \
                               LEFT JOIN genres ON tracks.GenreId = genres.GenreId \
                               LEFT JOIN invoice_items ON tracks.TrackId = invoice_items.TrackId \
                               GROUP BY genres.GenreId \
                               ORDER BY Total DESC",
                              conn,
                              index_col='GenreId')
```
