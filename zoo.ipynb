```python
import pandas as pd
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

data = pd.read_excel('zoo.xlsx')
```

```python
data
```

```python
data.isnull().sum()
```

```python
X = data.drop(['name','type'],axis=1)
y = data['type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### BAGGING MODEL

```python
bagging_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    n_estimators=50,
    max_samples=0.8,
    max_features=1,
    random_state=42,
)
```

```python
bagging_model.fit(X_train, y_train)
```

```python
y_pred = bagging_model.predict(X_test)
```

```python
bagging_akurasi = accuracy_score(y_test, y_pred)
bagging_akurasi
```

### ADA BOOST


```python
adaboost_model = AdaBoostClassifier(
    estimator = DecisionTreeClassifier(max_depth=1,random_state=42),
    n_estimators=50,
    learning_rate=1,
    random_state=42
)
```

```python
adaboost_model.fit(X_train, y_train)
```

```python
y_pred = adaboost_model.predict(X_test)
```

```python
adaboost_akurasi = accuracy_score(y_test, y_pred)
adaboost_akurasi
```

### GRADIEN BOOSTING

```python
gb_model = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
)
```

```python
gb_model.fit(X_train, y_train)
```

```python
y_pred = gb_model.predict(X_test)
```

```python
gb_akurasi = accuracy_score(y_test,y_pred)
gb_akurasi
```

```python
akurasi = {
    "Bagging": bagging_akurasi,
    "AdaBoost": adaboost_akurasi,
    "GradientBoosting": gb_akurasi
}

# Temukan model dengan akurasi terbaik
model_terbaik = max(akurasi, key=akurasi.get)
print(f"Model terbaik adalah {model_terbaik} dengan akurasi {akurasi[model_terbaik]:.2f}")
```
